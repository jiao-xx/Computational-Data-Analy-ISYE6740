# Holder Analysis

```python
holder_df = pd.read_csv('Unique_key_holder_detail_info.csv')
holder_df
```

```python
holder_df[['id', 'address', 'twitterUsername', 'ftUsername', 
       'lastOnline', 'lastMessageTime', 'holderCount', 'holdingCount',
       'followerCount', 'followingCount', 'shareSupply',
       'displayPrice', ]]
```

```python
holder_df['keysAvailable'] = holder_df['shareSupply'] - holder_df['holderCount']
```

```python
holder_df['keysAvailable']
```

## 1. Influence and Centrality Analysis:

Holder Count and Holding Count: Analyze the users with the highest holder counts (number of people holding their tokens) and holding counts (number of tokens they hold). This can help identify the most influential users in terms of token distribution.

```python
# Descriptive Statistics
desc_stats = holder_df[['holderCount', 'holdingCount', 'followerCount', 'followingCount']].describe()

desc_stats
```


### Descriptive Statistics:

- **holderCount**: 
  - Mean: 6.84
  - Users typically have 3.0 to 8.0 holders, but some have up to 120.0.
  - The distribution suggests a majority of users have relatively few holders, with a long tail of users having many.

- **holdingCount**: 
  - Mean: 27.08
  - The holding count varies significantly, with a high standard deviation (172.46), indicating a few users hold many tokens, skewing the average.
  - The majority of users hold between 4.0 to 17.0 tokens, but some hold as many as 7871.0.

- **followerCount**: 
  - Mean: 263.85
  - There is a wide range of follower counts, with some users having up to 69734.0 followers.
  - The large standard deviation (2383.67) indicates significant variability in follower counts.

- **followingCount**: 
  - Mean: 59.73
  - Users generally follow fewer accounts compared to their follower counts, with some following as many as 14856.0 accounts.
  - The distribution indicates most users follow a relatively small number of others, with a few exceptions.


## 2. Top users by Holding Count & Following Count

```python
# Top users by Holding Count
top_holding_count = holder_df.nlargest(10, 'holdingCount')

plt.figure(figsize=(12, 6))
bars = plt.bar(top_holding_count['twitterUsername'], top_holding_count['holdingCount'], color='blue')
plt.title('Top 10 Users by Holding Count')
plt.xlabel('Twitter Username')
plt.ylabel('Holding Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Add holding count values on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.0f}', 
                 xy=(bar.get_x() + bar.get_width() / 2, height), 
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points", 
                 ha='center', va='bottom')

plt.show()
```

```python
# Top users by Following Count
top_holding_count = holder_df.nlargest(10, 'followingCount')

plt.figure(figsize=(12, 6))
bars = plt.bar(top_holding_count['twitterUsername'], top_holding_count['followingCount'], color='red')
plt.title('Top 10 Users by Following Count')
plt.xlabel('Twitter Username')
plt.ylabel('Following Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Add holding count values on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.0f}', 
                 xy=(bar.get_x() + bar.get_width() / 2, height), 
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points", 
                 ha='center', va='bottom')

plt.show()
```

# Distributor Analysis

```python
# Load the dataset
file_path = 'Unique_key_distributors.csv'
data = pd.read_csv(file_path)

# Display the first few rows to understand the structure
data
```

```python
filtered_data = data[['id', 'address', 'twitterUsername', 'ftUsername', 'lastOnline',
       'lastMessageTime', 'holderCount', 'holdingCount', 
       'followerCount', 'followingCount', 'shareSupply', 'displayPrice']]

filtered_data
```

```python
filtered_data['displayPrice'] = data['displayPrice'] / 10**18
filtered_data['displayPrice'] 
```

## 1. Influence and Centrality Analysis:

Holder Count and Holding Count: Analyze the users with the highest holder counts (number of people holding their tokens) and holding counts (number of tokens they hold). This can help identify the most influential users in terms of token distribution.

```python
# Basic statistics for the key metrics
print(filtered_data[['holderCount', 'holdingCount', 'followerCount', 'followingCount']].describe())
```

## 2. Top users by Holder Count & Follower Count

```python
# Top users by Holder Count
top_holding_count = filtered_data.nlargest(10, 'holderCount')

plt.figure(figsize=(12, 6))
bars = plt.bar(top_holding_count['twitterUsername'], top_holding_count['holderCount'], color='blue')
plt.title('Top 10 Users by Holder Count')
plt.xlabel('Twitter Username')
plt.ylabel('Holder Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Add holding count values on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.0f}', 
                 xy=(bar.get_x() + bar.get_width() / 2, height), 
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points", 
                 ha='center', va='bottom')

plt.show()
```

```python
# Top users by Follower Count
top_holding_count = filtered_data.nlargest(10, 'followerCount')

plt.figure(figsize=(12, 6))
bars = plt.bar(top_holding_count['twitterUsername'], top_holding_count['followerCount'], color='Red')
plt.title('Top 10 Users by Follower Count')
plt.xlabel('Twitter Username')
plt.ylabel('Follower Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Add holding count values on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.0f}', 
                 xy=(bar.get_x() + bar.get_width() / 2, height), 
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points", 
                 ha='center', va='bottom')

plt.show()
```

```python
# Calculate Total Expected Return
filtered_data['totalExpectedReturn'] = filtered_data['shareSupply'] * filtered_data['displayPrice']
```

```python
top_total_return = filtered_data.nlargest(10, 'totalExpectedReturn')

# Create the plot
plt.figure(figsize=(12, 6))
bars = plt.bar(top_total_return['twitterUsername'], top_total_return['totalExpectedReturn'], color='purple')
plt.title('Top 10 Users by Total Expected Return')
plt.xlabel('Twitter Username')
plt.ylabel('Total Expected Return')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Add values on top of the bars
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.0f}', 
                 xy=(bar.get_x() + bar.get_width() / 2, height), 
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points", 
                 ha='center', va='bottom')

# Show the plot
plt.show()
```
